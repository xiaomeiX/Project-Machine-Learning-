### outlier

#### although the boxplot shows large amount of outlier, we decide to reserve those values

#### we use zscore to filter the data, and found those data may not be the outliers

```python
col = ['# of Delays.Carrier',
       '# of Delays.Late Aircraft', '# of Delays.National Aviation System',
       '# of Delays.Security', '# of Delays.Weather',
       'Carriers.Total', 'Flights.Cancelled',
       'Flights.Delayed', 'Flights.Diverted', 'Flights.On Time',
       'Flights.Total', 'Minutes Delayed.Carrier',
       'Minutes Delayed.Late Aircraft',
       'Minutes Delayed.National Aviation System', 'Minutes Delayed.Security',
       'Minutes Delayed.Total', 'Minutes Delayed.Weather']
```

```python
from scipy import stats

def deleteOutlier(dfBK, col):
    for attr in col:
        dfBK = dfBK[(np.abs(stats.zscore(dfBK[[attr]])) < 3)]
    return dfBK
```

#### discretization

```python
col = ['# of Delays.Carrier',
       '# of Delays.Late Aircraft', '# of Delays.National Aviation System',
       '# of Delays.Security', '# of Delays.Weather',
       'Carriers.Total', 'Flights.Cancelled',
       'Flights.Delayed', 'Flights.Diverted', 'Flights.On Time',
       'Flights.Total', 'Minutes Delayed.Carrier',
       'Minutes Delayed.Late Aircraft',
       'Minutes Delayed.National Aviation System', 'Minutes Delayed.Security',
       'Minutes Delayed.Total', 'Minutes Delayed.Weather']
```

```python
dfDiscreted = df.copy()
dfDiscreted = dfDiscreted.drop(['Rate'], axis = 1)
dfDiscreted.head()
```

```python
def getFiveNumber(dfCopy, attr):
    Q1 = dfCopy[attr].quantile(0.25)
    Q3 = dfCopy[attr].quantile(0.75)
    median =dfCopy[attr].median()
    return Q1, Q3, median
```

```python
for attr in col:

    Q1, Q3, median = getFiveNumber(dfDiscreted, attr)
    
    dfDiscreted.loc[dfDiscreted[attr] <= Q1, attr] = 0
    dfDiscreted.loc[(dfDiscreted[attr] > Q1) & (dfDiscreted[attr] <= median), attr] = 1
    dfDiscreted.loc[(dfDiscreted[attr] > median) & (dfDiscreted[attr] <= Q3), attr] = 2
    dfDiscreted.loc[(dfDiscreted[attr] > Q3), attr] = 3
 
```
